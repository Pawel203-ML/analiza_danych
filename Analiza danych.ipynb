# 2. Jakie rda pozyskania pracownika (Recruitment Source) s najlepsze, jeli zaley nam na jak najduszym stau pracownikw?

```python
#tworzenie kolumny ze stazem
df = data.copy() 

df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(365,'D')

df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
df
```

```python
order = df['RecruitmentSource'].value_counts(ascending=False).index
#SORTOWANIE wartosci do wyswietlania ujednoliconej kolejnosci w wykresach
df['RecruitmentSource'] = pd.Categorical(df['RecruitmentSource'], categories=order, ordered=True)
```

```python
#szukanie osob ze stazem wiekszym od 7 lat
df1 = df.copy()
df1 = df1.sort_values(by = 'Seniority')
df1 = df1[(df1['Seniority'] >= 7) & (df1['Seniority'] < 10)]
```

```python
#wykres ilosc osob zatrudnionyc hw dany sposob
sns.countplot(x = 'RecruitmentSource', data = df)
plt.xticks(rotation = 90)
plt.title('Ilosc osob zrekrutowanych danym sposobem')
plt.axhline(y = 20, color = 'red', linestyle = '-.', label = '>20 osob')
plt.axhline(y = 10, color = 'orange', linestyle = '-.', label = '>10 osob')
plt.legend(loc = (1.05, 0.9))
plt.show()
```

```python
sns.jointplot(x = 'Seniority', y = 'RecruitmentSource', data = df1)
for x in range(len(order)):
    plt.axhline(y = x + 0.5 , color = 'green', linestyle = '--', alpha = 0.3)
plt.title('Relacja pomiedzy sposobem zatrudnienia a strazem pracy od 7 do 10 lat', fontsize = 12, y = 1.2)
plt.show()
```

```python
#szukanie osob ze stazem wiekszym od 10 lat
df1 = df.sort_values(by = 'Seniority')
df1 = df1[df1['Seniority'] >= 10]
```

```python
sns.jointplot(x = 'Seniority', y = 'RecruitmentSource', data = df1)
for x in range(len(order)):
    plt.axhline(y = x + 0.5 , color = 'green', linestyle = '--', alpha = 0.3)
plt.title('Relacja pomiedzy sposobem zatrudnienia a strazem pracy > 10 lat', fontsize = 12, y = 1.2)
plt.show()
```

Wnioski: 
    Jesli zalezy nam na dugosci stazu pracy najlepszymi sposobami na to s: Billboard a pozniej Diversity Job Fair dla osob ktore osiagaja staz powyzej 10 lat
    natomiast dla osob ze stazem od 7 do 10 lat nalepszymi sposobami s: Search Engine - Google Bing Yahoo oraz Monster.com , a pozniej MBTA ads oraz Diversity Job Fair

# 4. Jak wyglda struktura wieku aktualnie zatrudnionych pracownikw?

```python
data['DOB'].head()
```

```python
age = data.copy()
age = age[['DOB', 'Employee_Name', 'SpecialProjectsCount']]
age['Age'] = age.apply(lambda row: -(row['DOB'] - dt.datetime(2019,9,27))//np.timedelta64(365,'D') , axis = 1)
age = age[age['Age'] > 0]
```

```python
sns.countplot(x = 'Age', data = age)
plt.axhline(y = 20, color = 'red', alpha = 0.5, linestyle = '--', label = '> 20 osob')
plt.axhline(y = 15, color = 'orange', alpha = 0.5, linestyle = '--', label = '> 15 osob')
plt.axhline(y = 10, color = 'violet', alpha = 0.5, linestyle = '--', label = '> 10 osob')
plt.legend(loc = (1.05, 0.8))
plt.title('Struktura wieku pracownikow obecnie pracujacych w firmie')
plt.xticks(rotation = 45)
plt.show()
```

Wnioski:
    Najwiecej pracownikow pracujacych obecnie w firmie jest z przedzialu od 31 do 36 lat oraz 40 lat
